```python
## Image Plate Diffraction System ##

## Image Loading and Plots ##

## Empa, Center for X-ray Analytics, D.Sapalidis, St. Gallen, Switzerland, 21.12.2023 ##

import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.ndimage import zoom

def load_and_display_image(image_path):
    # Read the image and convert it to a Numpy array
    image = np.array(Image.open(image_path))

    # Reproduce the Numpy array in grayscale
    grayscale_image = np.mean(image, axis=2)

    # Display the image with the Y-axis flipped
    plt.imshow(grayscale_image, cmap='gray', origin='lower')
    plt.title('Image in X, Y, Greyscale')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()

def display_interactive_intensity(image_path, zoom_factor=2):
    # Read the image and convert it to a Numpy array
    image = np.array(Image.open(image_path))

    # Reproduce the Numpy array in grayscale
    grayscale_image = np.mean(image, axis=2)

    # Increased resolution using zoom
    high_res_image = zoom(grayscale_image, zoom_factor)

    # Create arrays with coordinates and intensity values
    x_vals, y_vals = np.meshgrid(np.linspace(0, high_res_image.shape[1]-1, high_res_image.shape[1]),
                                 np.linspace(0, high_res_image.shape[0]-1, high_res_image.shape[0]))
    intensity_values = high_res_image.flatten()

    # 3D plot with Matplotlib
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(projection='3d')
    ax.plot_surface(x_vals, y_vals, high_res_image, cmap='gray', edgecolor='k')

    # Display the plot
    plt.show()

# Call the function for the main image with a zoom factor of 2
main_image_path = 'C:/Users/sapd/Desktop/X23_0035-Pigment UPDATE/Pigm5-IPDS-Annealing Effect, 19-20.12.2023/Pigm5-IPDS-Images/DCSQI_annealed_40mm_240min_ω0_φ10rpm_201223.tif'
load_and_display_image(main_image_path)
display_interactive_intensity(main_image_path, zoom_factor=2)
```
